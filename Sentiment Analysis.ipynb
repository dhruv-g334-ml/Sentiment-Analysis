# To count different words in a sentence

```python
#to count the different words in a sentence as done by turicreate
d = dict()
p = products['review'][0].lower()
for i in p.split(" "):
    d[i] = d.get(i,0)+1
d
```

```python
def count(x):
    d = dict()
    y = str(x).lower()
    for i in y.split(" "):
        d[i] = d.get(i,0)+1
    return d
        
```

```python
products['word_count'] = products['review'].apply(count)
```

# Training the model

```python
x = products['word_count']
y = products['sentiment']
```

The data is from 500 head entries only.

### To convert dictionary to correct format 

To convert the dictionary format to a format compatible with the ML model, we use the library as below.

```python
from sklearn.feature_extraction import DictVectorizer

dictVectorizer = DictVectorizer()

x = dictVectorizer.fit_transform(x)
```

### Splitting the data into train and test sets

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y, random_state = 0)
```

### Training the model

Since the dataset is large, we require more time to train the model

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression(max_iter = 100000).fit(x_train,y_train)

y_pred = lgr.predict_proba(x)
```

Now, here we mentioned the predict_proba() instead of predict(). This predict_proba() indicates the probability prediction of 0 or 1 in a case of output. Here, the first term shows the probability of output of 0 and second term of array shows the probability of output 1. So, in our model, we usually use the probability of output 1.
